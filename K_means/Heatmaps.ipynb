**什么是 heatmap**

Heatmap 是一种用 颜色深浅来表示数值大小 的二维图表。

横轴、纵轴通常是变量，颜色表示对应位置的数值。

常用于 相关性矩阵、聚类结果可视化、地理分布 等场景。

颜色深 → 数值大

颜色浅 → 数值小

一眼能看出 模式、相关性、分布差异
