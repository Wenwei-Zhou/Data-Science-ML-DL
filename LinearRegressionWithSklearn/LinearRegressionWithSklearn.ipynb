# **Multiple linear regression**

```python
dataMul = pd.read_csv('sample_data/1.02.+Multiple+linear+regression.csv')
dataMul.head()
```

```python
dataMul.describe()
```

# **Create multiple linear regression**

**Declear the dependent and independent variable**

```python
xMul = dataMul[['SAT','Rand 1,2,3']]
yMul = dataMul['GPA']
```

**Regression itself**

```python
reg = LinearRegression()
reg.fit(xMul,yMul)
```

**SAT coef, Rand1,2,3 coef**

```python
reg.coef_
```

**const coef**

```python
reg.intercept_
```

# **Calculating the R-squared**

**R-squared：决定系数，表示模型能解释多少 GPA 的波动**

```python
reg.score(xMul,yMul)
```

SAT 能解释 GPA 波动的 40.6%，剩下的 59.4% 差异，需要其他因素来解释（比如努力程度、课程选择、学习环境等）。

**Formula for Adjusted R^2**

R²adj = 1 - [(1 - R²)(n - 1) / (n - p - 1)]

```python
xMul.shape
```

```python
RSquared = reg.score(xMul,yMul)

n = xMul.shape[0]
p = xMul.shape[1]

adjusted_r2 = 1-(1 - RSquared)(n - 1)/(n - p - 1)
adjusted_r2
```

# **Feature selection**

我们已经创建了一个multiple regression, 并计算了其调整后的R-squared｡

R-squared的值表明两个变量中的一个并不能增加回归的解释力｡ (SAT和rand1,2,3分别获得的R-squared, rand1,2,3的R-squared表明rand1,2,3这个数据模型没有用)

因此, 如果我们能识别出是哪一个并将其删除, 我们就可以进一步简化模型｡

```python
from sklearn.feature_selection import f_regression
```

```python
f_regression(xMul,yMul)
```

```python
p_values = f_regression(xMul,yMul)[1]
p_values
```

```python
p_values.round(3)
```

**F-statistic 和 p-value 的关系**

F-statistic 测试的是：这个特征对 y 的解释力是否显著大于 0。

p-value 是根据 F-statistic 算出来的显著性概率。

**如果特征:**

p-value 很小（比如 < 0.05），说明它对 y 的影响显著 → 这个特征重要

p-value 大（比如 > 0.5），说明它对 y 的解释力不显著 → 可以考虑剔除

# **Creating a summary table**

```python
reg_summary = pd.DataFrame(data = xMul.columns.values, columns=['Features'])
reg_summary
```

```python
reg_summary['coefficients'] = reg.coef_
reg_summary['p-values'] = p_values.round(3)
```

```python
reg_summary
```

**coefficients：回归系数，代表这个特征对目标变量（如 GPA）的影响大小和方向。**

SAT 的系数 0.001654 → SAT 每提高 1 分，GPA 平均提高 0.001654。

Rand 的系数 -0.008270 → 随机数对 GPA 的影响是负的，但很小。

**p-values：检验该特征对模型是否有显著贡献（假设检验）。**

SAT 的 p = 0.000 → 显著，说明 SAT 对 GPA 的预测有意义。

Rand 的 p = 0.676 → 不显著，说明随机数和 GPA 没关系。

**coefficients和p-values都说明Rand1,2,3没有用**

# **Standardization**

fit()：学到每一列的均值和标准差

transform()：应用标准化，把原始数据变成：

x′=（x−μ​）/ σ

这样每个特征的均值=0，标准差=1。

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
# 1. 定义一个标准化工具
```

```python
scaler.fit(xMul)
# 2. 计算每个特征的均值和标准差
```

```python
x_scaled = scaler.transform(xMul)
# 3. 用均值和标准差把数据转换
```

```python
x_scaled
```

# **Regression with scaled features**

```python
reg = LinearRegression()
reg.fit(x_scaled,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

# **Creating a summary table**

```python
reg_summary = pd.DataFrame([['Bias'],['SAT'],['Rand 1,2,3']], columns=['Features'])
reg_summary['weights'] = reg.intercept_, reg.coef_[0], reg.coef_[1]
```

```python
reg_summary
```

SAT 的 weight = 0.0016 → SAT 每提高 1 分，GPA 提高 0.0016

Rand 的 weight = -0.008 → Rand 越大，GPA 越低（Rand对我们的输出几乎没有贡献）

bias = 0.27 → 截距，当 SAT 和 Rand=0 时，预测 GPA=0.27
