# **Market segmentation example**

# **Import the relevant libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

# **Load the data**

```python
data = pd.read_csv('sample_data/3.12.+Example.csv')
```

```python
data
```

# **Plot that data**

```python
plt.scatter(data['Satisfaction'],data['Loyalty'])
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

# **Select the features**

```python
x = data.copy()
```

# **Clustering**

```python
kmeans = KMeans(2)
kmeans.fit(x)
```

# **Clusting results**

```python
clusters = x.copy()
clusters['cluster_pred']=kmeans.fit_predict(x)
```

```python
plt.scatter(clusters['Satisfaction'],clusters['Loyalty'],c=clusters['cluster_pred'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

# **Stadardize the variable**

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
x_scaled
```

**preprocessing.scale(x)**

标准化，将数据无论是多大的数字都变成特征的均值=0，标准差=1。ex：（-2，-1，0，1，2）

# **Take advantage of the Elbow method**

```python
wcss = []

for i in range(1,10):
  kmeans = KMeans(i)
  kmeans.fit(x_scaled)
  wcss_iter = kmeans.inertia_
  wcss.append(wcss_iter)

wcss
```

**for i in range(1,10):**

n_clusters从1到10，每个执行一遍

**kmeans = KMeans(i)**

执行KMeans从1到10，每个执行一遍

**kmeans.fit(x_scaled)**

用标准化后的数据进行fit()

**wcss_iter = kmeans.inertia_**

用inertia_算出紧密度

**wcss.append(wcss_iter)**

n_clusters1-10,每执行一个得出inertia，把output(紧密度)放进wcss[]里面

```python
plt.plot(range(1,10),wcss) # x=range(1,10) y=wcss
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# **Explore clustering solutions and select the number of clusters**

```python
kmeans_new = KMeans(4)
kmeans_new.fit(x_scaled)
clusters_new = x.copy()
clusters_new['cluster_pred'] = kmeans_new.fit_predict(x_scaled)
```

```python
clusters_new
```

```python
plt.scatter(clusters_new['Satisfaction'],clusters_new['Loyalty'],c=clusters_new['cluster_pred'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

**散射图现实n_clusters=4是最合适的，将数据分成四个类聚：**

左下角：satisfaction low, loyalty low

左上角：satisfaction low, loyalty high

右下角：satisfaction high, loyalty low

左上角：satisfaction high, loyalty high
